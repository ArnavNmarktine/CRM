```python
def generate_random_datetime(start_date, end_date):
    time_delta = end_date - start_date
    random_days = random.randint(0, time_delta.days)
    random_seconds = random.randint(0, 86400)  # Total seconds in a day (24 hours * 60 minutes * 60 seconds)
    return start_date + timedelta(days=random_days, seconds=random_seconds)

# Define the range of dates for random date-time generation
start_date_str = "2021-01-01"
end_date_str = "2022-12-31"

start_date = datetime.strptime(start_date_str, "%Y-%m-%d")
end_date = datetime.strptime(end_date_str, "%Y-%m-%d")

# Generate 10 random date-time values
random_date_times = [generate_random_datetime(start_date, end_date) for _ in range(10)]

# Display the results
for dt in random_date_times:
    print(dt.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
def split_amount_unequally(total_amount, num_elements):
    # Generate random ratios
    ratios = [random.random() for _ in range(num_elements)]

    # Calculate the total ratio sum
    total_ratio = sum(ratios)

    # Initialize the list to store split amounts
    split_amounts = []

    # Calculate and append the split amount for each element
    for ratio in ratios:
        split_amount = total_amount * (ratio / total_ratio)
        split_amounts.append(split_amount)

    return split_amounts

# # Example usage:
# total_amount = 1000
# num_elements = 5
# result = split_amount_unequally(total_amount, num_elements)
# print(result)
```

```python
def normal_distribution(x, mean, std_dev):
    return 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Define the parameters of the normal distribution
mean = 0     # Mean (center of the distribution)
std_dev = 1  # Standard deviation (spread of the distribution)

# Generate x-values (for example, from -5 to 5)
x_values = np.linspace(-5, 5, 100)

# Calculate the corresponding probabilities using the normal_distribution function
probabilities = normal_distribution(x_values, mean, std_dev)

# Create the plot
plt.plot(x_values, probabilities, label="Normal Distribution")
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def generate_fake_sales_data(num_days, initial_sales=random.randint(1000,2000),growth_rate=0.005 , seasonal_variation= 0.1 ):
    # Generate dates
    dates = pd.date_range(start='2023-01-01', periods=num_days, freq='D')


    # Generate the base sales data with exponential growth
    growth_factor = (1 + growth_rate) ** np.arange(num_days)
    sales_data = initial_sales * growth_factor

    # Add seasonal variations
    seasonal_factor = 1 + seasonal_variation * np.sin(2 * np.pi * np.arange(num_days) / 30)
    sales_data *= seasonal_factor

    # Add random noise to the data
    np.random.seed(42)  # For reproducibility
    noise = np.random.normal(scale=0.05, size=num_days)  # Adjust the scale for desired noise level
    sales_data += noise * sales_data

    # Create a DataFrame to store the data
    df = pd.DataFrame({'Date': dates, 'Sales': sales_data})

    return df



# Generate the fake sales data
sales_data = generate_fake_sales_data(365*3)

# Plot the line graph
plt.figure(figsize=(12, 6))
plt.plot(sales_data['Date'], sales_data['Sales'], label='Sales')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.title('Fake Sales Data')
plt.legend()
plt.grid(True)
plt.show()
#generate_fake_sales_data(365)['Sales'].to_list()
```

```python

products = ['Vriti', 'Generative_AI_Solutions', 'Vriddhi', 'UnifID_IAM', 'Parse_Connect']
regions = ['North', 'South', 'East', 'West']

def generate_fake_data():
    data = []
    products_seen = set()

    for product in products:
        if product not in products_seen:
            region = random.choice(regions)

            for i in regions:
              no_of_salesmen = random.randint(1, 5)
              data.append((product, i, no_of_salesmen))
            products_seen.add(product)

    return data

# Generate fake data with one data point per product
fake_data = generate_fake_data()

# Create a DataFrame from the fake data
df = pd.DataFrame(fake_data, columns=['product', 'region', 'no_of_salesmen'])

# Print the DataFrame

# df['region_wise_unique_name']=df['product']+'-'+df['region']
# df.sort_values(by=['product', 'region'], inplace=True)
df.to_excel("df.xlsx")
```
